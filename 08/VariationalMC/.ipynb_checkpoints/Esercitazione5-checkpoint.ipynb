# Esercizio 5.1

L'esercizio richiede di utilizzare l'algoritmo di Metropolis per campionare le distribuzioni di probabilità data dal modulo quadro delle autofunzioni delle configurazioni 1s e 2p dell'atomo di idrogeno al fine di stimare $\langle r \rangle$ con relativa incertezza per entrambe le configurazioni

Nella prima parte dell'esercizio utilizzo una funzione di transizione uniforme 

$$
T(\vec{x}|\vec{y}) = \cases{1/a^3, \hspace{10pt} |x_1 - y_1| < a/2 \land |x_2 - y_2| < a/2 \land |x_3 - y_3| < a/2 \\
0, \hspace{10pt} \text{otherwise}}
$$

il parametro a è stato scelto in modo tale da ottenere un tasso di accettazione prossimo al 50% per entrambe le distribuzioni di probabilità da campionare

```python
from IPython.display import display, Latex
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

ru, au1, au2 = np.loadtxt('raggiuni.dat', usecols=(0,1,2), delimiter=" ", unpack=True)
plt.plot(ru, au1, label='AR(1s)')
plt.plot(ru, au2, label='AR(2p)')
au1_b = np.abs(au1-0.5)
au2_b = np.abs(au2-0.5)
plt.plot(ru, au1_b+au2_b, label='|AR(1s)-0.5|+|AR(2p)-0.5|')
plt.xlabel('a')
plt.legend()
plt.show()
bmin = np.amin(au1_b+au2_b)
imin = np.where(au1_b+au2_b == bmin)
minr = ru[imin]
print("La migliore scelta per il parametro a è " + str(minr) + "\n")
print("con una accettazione nelle due configurazioni del " + str(au1[imin]) + " e " + str(au2[imin]))
```

## Risultati ottenuti con funzione di transizione uniforme

### Configurazione 1s
In questa simulazione ho utilizzato l'algoritmo di metropolis per campionare la distribuzione di probabilità data dal modulo quadro dell'autofunzione 1s dell'atomo di idrogeno 

$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$

la simulazione viene eseguita utilizzando $a_0 = 1$ e i valori di $\langle r \rangle$ ottenuti sono dunque in unità di $a_0$.
Il valore aspettato dai calcoli analitici è
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
$$
per questa simulazione ho eseguito $10^7$ estrazioni suddivise in 100 blocchi. Prima di utilizzare i valori estratti con l'algoritmo di Metropolis per la simulazione ho eseguito 100 estrazioni "a vuoto" al fine di raggiungere la condizione di convergenza desiderata.

```python
r, err = np.loadtxt('uni1s', usecols=(0,1), delimiter=" ", unpack=True)
x = np.arange(0, len(r))
plt.errorbar(x, r-1.5, err)
plt.xlabel("block")
plt.ylabel('$\\langle r \\rangle /a_0 -1.5$')

plt.grid()
plt.show()

print("stima finale di r/a_0: " + str(r[-1]) )
print("incertezza: " + str(err[-1]))
```

### Configurazione 2p
in questa simulazione ho invece campionato la distribuzione di probabilità ottenuta dal modulo quadro dell'autofunzione 2p dell'atomo di idrogeno

$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$

anche in questo caso il risultato di $\langle r \rangle$ ottenuto si intende espresso in unità di $a_0$, il risultato aspettato dai calcoli analitici è 


```python
r, err = np.loadtxt('uni2p', usecols=(0,1), delimiter=" ", unpack=True)
x = np.arange(0, len(r))
plt.errorbar(x, r-5, err)
plt.xlabel("block")
plt.ylabel('$\\langle r \\rangle /a_0 -5$')
plt.grid()
plt.show()

print("stima finale di r/a_0: " + str(r[-1]) )
print("incertezza: " + str(err[-1]))
```

Per entrambe le distribuzioni di probabilità campionate ho anche eseguito delle simulazioni salvando tutti i punti estratti. I risultati sono i seguenti e mostrano il successo dell'algoritmo di metropolis nel campionamento delle distribuzioni di probabilità scelte

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=[10,5])

x, y, z = np.loadtxt('points1S', usecols=(0,1,2), delimiter=" ", unpack=True)
ax1s = fig.add_subplot(1, 2, 1, projection='3d', label='1s')
ax1s.scatter(x, y, z, c=z, marker='.')
ax1s.set_xlabel('x')
ax1s.set_ylabel('y')
ax1s.set_zlabel('z')
ax1s.set_title('Stato 1s')
ax1s.view_init(10, 30)

x, y, z = np.loadtxt('points2P', usecols=(0,1,2), delimiter=" ", unpack=True)
ax2p = fig.add_subplot(1, 2, 2, projection='3d', label='2p')
ax2p.scatter(x, y, z, c=z, marker='.')
ax2p.set_xlabel('x')
ax2p.set_ylabel('y')
ax2p.set_zlabel('z')
ax2p.set_title('Stato 2p')
ax2p.view_init(10, 30)

plt.show()
```

Per eseguire queste simulazioni ho utilizzato come punto iniziale per l'algoritmo di Metropolis l'origine. Ho eseguito anche delle simulazioni impostando come punto iniziale (1000, 1000, 1000), punto in cui le distribuzioni di probabilità campionate assumono valori prossimi a 0. Mantenendo la stessa lunghezza della fase di equilibrazione (100 estrazioni) ho ottenuto dei punti distribuiti nello spazio nel seguente modo

```python
fig = plt.figure(figsize=[10,5])

x, y, z = np.loadtxt('points1SFAR', usecols=(0,1,2), delimiter=" ", unpack=True)
ax1s = fig.add_subplot(1, 2, 1, projection='3d', label='1s')
ax1s.scatter(x, y, z, c=z, marker='.')
ax1s.set_xlabel('x')
ax1s.set_ylabel('y')
ax1s.set_zlabel('z')
ax1s.set_title('Stato 1s')
ax1s.view_init(10, 30)

x, y, z = np.loadtxt('points2PFAR', usecols=(0,1,2), delimiter=" ", unpack=True)
ax2p = fig.add_subplot(1, 2, 2, projection='3d', label='2p')
ax2p.scatter(x, y, z, c=z, marker='.')
ax2p.set_xlabel('x')
ax2p.set_ylabel('y')
ax2p.set_zlabel('z')
ax2p.set_title('Stato 2p')
ax2p.view_init(10, 30)

plt.show()
```

Si può osservare che facendo partire l'algoritmo da un punto molto lontano dall'origine, e quindi dalla regione in cui la distribuzione di probabilità che si vuole campionare è significativa, non si riesce a raggiungere la convergenza

## Risultati ottenuti con una distribuzione normale multivariata come funzione di transizione

In questa parte dell'esercizio ho eseguito le stesse simulazioni utilizzando come funzione di transizione $T(\vec{x}|\vec{y})$ una distribuzione normale multivariata centrata in $\vec{y}$. 
Il parametro di questa distribuzione è la deviazione standard della distribuzione. Ho eseguito una simulazione analoga a quella svolta nel caso precedente per valutare come varia il tasso di accettazione dell'algoritmo di metropolis al variare di questo parametro.

```python
ru, au1, au2 = np.loadtxt('raggigauss.dat', usecols=(0,1,2), delimiter=" ", unpack=True)
plt.plot(ru, au1, label='AR(1s)')
plt.plot(ru, au2, label='AR(2p)')
au1_b = np.abs(au1-0.5)
au2_b = np.abs(au2-0.5)
plt.plot(ru, au1_b+au2_b, label='|AR(1s)-0.5|+|AR(2p)-0.5|')
plt.xlabel('a')
plt.legend()
plt.show()
bmin = np.amin(au1_b+au2_b)
imin = np.where(au1_b+au2_b == bmin)
minr = ru[imin]
print("La migliore scelta per il parametro a è " + str(minr) + "\n")
print("con una accettazione nelle due configurazioni del " + str(au1[imin]) + " e " + str(au2[imin]))
```
